Here's the Manim code. It creates a scene with a circle and a line, positions them so they intersect, and then animates their appearance.

````python
from manim import *

class CircleAndLineIntersection(Scene):
    def construct(self):
        # Create a circle
        circle = Circle(radius=1.5, color=BLUE)
        circle.shift(LEFT * 0.5)

        # Create a line
        line = Line(start=3 * LEFT + UP, end=3 * RIGHT + DOWN, color=RED)

        # Add a title
        title = Text("Circle and Line Intersection").to_edge(UP)

        # Animate the creation of the circle and line
        self.play(Write(title))
        self.play(Create(circle))
        self.play(Create(line))
        self.wait(2)
````

Here's a new Manim scene for a dynamic visualization of a circle expanding from the origin, intersecting a fixed line. It will display the circle's equation and the intersection coordinates in real-time.

````python
from manim import *
import numpy as np

class CircleLineIntersectionDynamic(Scene):
    def construct(self):
        # Title
        title = Text("Dynamic Circle-Line Intersection").to_edge(UP)
        self.play(Write(title))
        self.wait(0.5)

        # 1. Line
        # The line passes through (-3, 1) and (3, -1), its equation is y = -x/3
        line = Line(start=3 * LEFT + UP, end=3 * RIGHT + DOWN, color=RED)
        self.play(Create(line))

        # 2. Circle with increasing radius from origin
        radius_tracker = ValueTracker(0.01) # Start with a very small radius
        circle = always_redraw(
            lambda: Circle(radius=radius_tracker.get_value(), color=BLUE).move_to(ORIGIN)
        )
        self.add(circle)

        # 3. Equation of the circle
        circle_equation_label = MathTex("Circle: ", color=BLUE).to_corner(DOWN + LEFT)
        circle_equation_formula = always_redraw(
            lambda: MathTex(f"x^2 + y^2 = {radius_tracker.get_value()**2:.2f}", color=BLUE)
                    .next_to(circle_equation_label, RIGHT, buff=0.2)
        )
        self.play(Write(circle_equation_label), FadeIn(circle_equation_formula))

        # 4. Line equation
        line_equation_label = MathTex("Line: ", color=RED).next_to(circle_equation_label, UP, aligned_edge=LEFT)
        line_equation_formula = MathTex("y = -\frac{1}{3}x", color=RED).next_to(line_equation_label, RIGHT, buff=0.2)
        self.play(Write(line_equation_label), Write(line_equation_formula))
        self.wait(0.5)

        # 5. Intersection coordinates
        intersection_label = MathTex("Intersection Points:", color=GREEN).to_corner(UP + RIGHT)
        self.add(intersection_label)

        def get_intersection_mobjects():
            r_val = radius_tracker.get_value()

            # For the line y = -x/3 and circle x^2 + y^2 = r^2
            # Intersection points are: ( +/- 3r / sqrt(10), +/- r / sqrt(10) )
            sqrt_10 = np.sqrt(10)

            # Point 1 (for positive x and negative y)
            x1 = (3 * r_val) / sqrt_10
            y1 = (-r_val) / sqrt_10

            # Point 2 (for negative x and positive y)
            x2 = (-3 * r_val) / sqrt_10
            y2 = (r_val) / sqrt_10

            point1_dot = Dot(np.array([x1, y1, 0]), color=GREEN)
            point2_dot = Dot(np.array([x2, y2, 0]), color=GREEN)

            coords_text_1 = MathTex(
                f"P_1 = ({x1:.2f}, {y1:.2f})", color=GREEN
            ).next_to(intersection_label, DOWN, aligned_edge=RIGHT)
            coords_text_2 = MathTex(
                f"P_2 = ({x2:.2f}, {y2:.2f})", color=GREEN
            ).next_to(coords_text_1, DOWN, aligned_edge=RIGHT)

            return VGroup(point1_dot, point2_dot, coords_text_1, coords_text_2)

        intersection_display = always_redraw(get_intersection_mobjects)
        self.add(intersection_display)

        # Animate radius increasing
        self.play(
            radius_tracker.animate.set_value(5), # Increase radius to 5
            run_time=10,
            rate_func=linear
        )
        self.wait(2)
````

````python
from manim import *

class LinearRegressionAnimation(Scene):
    def construct(self):
        # Title
        title = Text("Linear Regression (Visual Intuition)", font_size=36)
        self.play(Write(title))
        self.wait(1)
        self.play(title.animate.to_edge(UP))

        # Axes
        axes = Axes(
            x_range=[0, 10, 1],
            y_range=[0, 10, 1],
            x_length=8,
            y_length=5,
            axis_config={"include_numbers": True}
        ).to_edge(DOWN)

        labels = axes.get_axis_labels(x_label="X (Study Hours)", y_label="Y (Exam Score)")
        self.play(Create(axes), Write(labels))

        # Data points (example dataset)
        points_coords = [(1,2), (2,3), (3,5), (4,4), (5,6), (6,7), (7,8)]
        dots = VGroup(*[
            Dot(axes.coords_to_point(x, y), color=BLUE)
            for x, y in points_coords
        ])

        self.play(FadeIn(dots, shift=UP))
        data_label = Text("Data Points", font_size=24).next_to(axes, RIGHT)
        self.play(Write(data_label))
        self.wait(1)

        # Initial random line
        line = axes.plot(lambda x: 0.5*x + 1, color=YELLOW)
        line_label = Text("Regression Line", font_size=24, color=YELLOW).next_to(line, UP)

        self.play(Create(line), Write(line_label))
        self.wait(1)

        # Animate line improving (simulate fitting)
        better_line = axes.plot(lambda x: 0.9*x + 1, color=YELLOW)
        best_line = axes.plot(lambda x: 1.0*x + 1, color=GREEN)

        self.play(Transform(line, better_line))
        self.wait(1)
        self.play(Transform(line, best_line), line_label.animate.set_color(GREEN))
        self.wait(1)

        # Residual lines (errors)
        residuals = VGroup()
        for x, y in points_coords:
            predicted_y = x + 1  # matches best_line: y = x + 1
            start = axes.coords_to_point(x, y)
            end = axes.coords_to_point(x, predicted_y)
            residuals.add(Line(start, end, color=RED))

        error_label = Text("Error (Residuals)", font_size=24, color=RED).next_to(axes, LEFT)

        self.play(Create(residuals), Write(error_label))
        self.wait(1)

        # Shrink residuals to show minimization
        self.play(*[r.animate.scale(0.3, about_point=r.get_start()) for r in residuals])
        self.wait(1)

        # Show error value decreasing
        error_text = Text("Total Error ↓", font_size=28, color=RED).to_edge(RIGHT)
        self.play(Write(error_text))
        self.wait(2)

        # Prediction example
        x_val = 8
        y_pred = x_val + 1

        x_point = axes.coords_to_point(x_val, 0)
        pred_point = axes.coords_to_point(x_val, y_pred)

        vertical_line = Line(x_point, pred_point, color=WHITE)
        pred_dot = Dot(pred_point, color=GREEN)

        pred_label = Text("Prediction", font_size=24).next_to(pred_dot, RIGHT)

        self.play(Create(vertical_line), FadeIn(pred_dot))
        self.play(Write(pred_label))
        self.wait(2)

        # Final highlight
        final_text = Text("Best Fit Line Ready!", font_size=30, color=GREEN).to_edge(DOWN)
        self.play(Write(final_text))

        self.wait(3)
````

````python
from manim import *
import numpy as np

class KNNAnimation(Scene):
    def construct(self):
        # Title
        title = Text("K-Nearest Neighbors (KNN)", font_size=36)
        self.play(Write(title))
        self.wait(1)
        self.play(title.animate.to_edge(UP))

        # Axes
        axes = Axes(
            x_range=[0, 10, 1],
            y_range=[0, 10, 1],
            x_length=8,
            y_length=5,
            axis_config={"include_numbers": True}
        ).to_edge(DOWN)

        labels = axes.get_axis_labels(
            Text("Feature 1"),
            Text("Feature 2")
        )

        self.play(Create(axes), Write(labels))

        # Dataset (2 classes)
        class_A = [(2,3), (3,3), (3,4), (4,3)]
        class_B = [(6,6), (7,7), (8,6), (7,5)]

        dots_A = VGroup(*[
            Dot(axes.coords_to_point(x,y), color=BLUE)
            for x,y in class_A
        ])

        dots_B = VGroup(*[
            Dot(axes.coords_to_point(x,y), color=RED)
            for x,y in class_B
        ])

        label_A = Text("Class A", font_size=24, color=BLUE).next_to(axes, RIGHT)
        label_B = Text("Class B", font_size=24, color=RED).next_to(label_A, DOWN)

        self.play(FadeIn(dots_A), FadeIn(dots_B))
        self.play(Write(label_A), Write(label_B))
        self.wait(1)

        # New query point
        query_coords = (5,5)
        query_dot = Dot(axes.coords_to_point(*query_coords), color=YELLOW)
        query_label = Text("New Point", font_size=24).next_to(query_dot, UP)

        self.play(FadeIn(query_dot, scale=1.5))
        self.play(Write(query_label))
        self.wait(1)

        # Combine all points
        all_points = class_A + class_B
        all_dots = list(dots_A) + list(dots_B)

        # Draw distance lines
        lines = VGroup()
        distances = []

        for (x,y), dot in zip(all_points, all_dots):
            line = Line(
                query_dot.get_center(),
                dot.get_center(),
                color=GRAY
            )
            lines.add(line)

            dist = np.linalg.norm(np.array(query_coords) - np.array([x,y]))
            distances.append((dist, dot, line))

        self.play(Create(lines))
        self.wait(1)

        # Sort by distance
        distances.sort(key=lambda x: x[0])

        # Choose K = 3
        k = 3
        nearest = distances[:k]

        # Highlight nearest neighbors
        highlight_group = VGroup()

        for _, dot, line in nearest:
            highlight_group.add(dot, line)

        self.play(
            *[obj.animate.set_color(GREEN) for obj in highlight_group]
        )

        k_label = Text("K = 3 nearest neighbors", font_size=24, color=GREEN).to_edge(RIGHT)
        self.play(Write(k_label))
        self.wait(2)

        # Voting
        votes_A = sum(1 for _, d, _ in nearest if d in dots_A)
        votes_B = sum(1 for _, d, _ in nearest if d in dots_B)

        if votes_A > votes_B:
            result_color = BLUE
            result_text = "Class A"
        else:
            result_color = RED
            result_text = "Class B"

        result = Text(f"Classified as: {result_text}", font_size=30, color=result_color)
        result.to_edge(DOWN)

        self.play(Write(result))

        # Animate query point changing color
        self.play(query_dot.animate.set_color(result_color))

        self.wait(3)
````

````python
from manim import *

class QuadraticDiscriminant(Scene):
    def construct(self):
        # Title
        title = Text("Quadratic Equation & Discriminant", font_size=36)
        self.play(Write(title))
        self.wait(1)
        self.play(title.animate.to_edge(UP))

        # Axes
        axes = Axes(
            x_range=[-5, 5, 1],
            y_range=[-5, 10, 1],
            x_length=8,
            y_length=5,
            axis_config={"include_numbers": True}
        ).to_edge(DOWN)

        labels = axes.get_axis_labels(Text("x"), Text("y"))
        self.play(Create(axes), Write(labels))

        # Initial quadratic (D > 0)
        a, b, c = 1, -2, -3  # D = 16 > 0

        graph = axes.plot(lambda x: a*x**2 + b*x + c, color=BLUE)

        equation = MathTex("ax^2 + bx + c = 0").to_edge(LEFT)
        self.play(Create(graph), Write(equation))
        self.wait(1)

        # Roots visualization
        roots = [-1, 3]
        root_dots = VGroup(*[
            Dot(axes.coords_to_point(r, 0), color=YELLOW)
            for r in roots
        ])

        root_label = Text("Roots (x-intercepts)", font_size=24).next_to(axes, RIGHT)

        self.play(FadeIn(root_dots), Write(root_label))
        self.wait(2)

        # Show quadratic formula
        formula = MathTex(
            "x = \\frac{-b \\pm \\sqrt{b^2 - 4ac}}{2a}"
        ).scale(0.8).next_to(equation, DOWN)

        self.play(Write(formula))
        self.wait(2)

        # Highlight discriminant
        discrim = MathTex("D = b^2 - 4ac").set_color(YELLOW)
        discrim.next_to(formula, DOWN)

        self.play(Write(discrim))
        self.wait(2)

        # CASE 1: D > 0
        case1 = Text("D > 0 : Two real roots", font_size=24, color=GREEN).to_edge(RIGHT)
        self.play(Write(case1))
        self.wait(2)

        # CASE 2: D = 0
        new_graph = axes.plot(lambda x: x**2 - 2*x + 1, color=ORANGE)  # (x-1)^2
        new_dot = Dot(axes.coords_to_point(1, 0), color=YELLOW)

        case2 = Text("D = 0 : One repeated root", font_size=24, color=ORANGE).to_edge(RIGHT)

        self.play(
            Transform(graph, new_graph),
            FadeOut(root_dots),
            Transform(case1, case2)
        )
        self.play(FadeIn(new_dot))
        self.wait(2)

        # CASE 3: D < 0
        new_graph2 = axes.plot(lambda x: x**2 + 1, color=RED)

        case3 = Text("D < 0 : No real roots", font_size=24, color=RED).to_edge(RIGHT)

        self.play(
            Transform(graph, new_graph2),
            FadeOut(new_dot),
            Transform(case1, case3)
        )
        self.wait(2)

        # Linear algebra perspective
        la_text = Text("y = ax² + bx + c = combination of basis functions", font_size=24)
        la_text.to_edge(LEFT)

        self.play(Write(la_text))
        self.wait(2)

        basis = MathTex("x^2,", "\\quad x,", "\\quad 1").next_to(la_text, DOWN)
        self.play(Write(basis))
        self.wait(2)

        # Final summary
        summary = VGroup(
            Text("Roots = Intersection with X-axis", font_size=24),
            Text("Discriminant decides number of roots", font_size=24)
        ).arrange(DOWN).to_edge(DOWN)

        self.play(Write(summary))
        self.wait(3)
````

````python
from manim import *

class SkywardStory(Scene):
    def construct(self):

        # -----------------------------
        # Scene 1: City Introduction
        # -----------------------------
        buildings = VGroup(*[
            Rectangle(width=0.5, height=h, fill_opacity=1, color=BLUE_D)
            for h in [2, 3, 1.5, 2.5, 3.5, 2]
        ]).arrange(RIGHT, buff=0.2).to_edge(DOWN)

        self.play(FadeIn(buildings))
        self.play(self.camera.frame.animate.shift(RIGHT * 1.5), run_time=2)

        text1 = Text("A city that never stops...", font_size=28).to_edge(UP)
        self.play(Write(text1))
        self.wait(2)
        self.play(FadeOut(text1))

        # -----------------------------
        # Scene 2: Aarav (Mechanic)
        # -----------------------------
        mechanic = Circle(radius=0.3, color=YELLOW).shift(LEFT * 3 + DOWN)
        tool = Line(mechanic.get_right(), mechanic.get_right() + RIGHT * 0.5)

        self.play(FadeIn(mechanic, tool))

        # simple "working" motion
        for _ in range(3):
            self.play(tool.animate.rotate(0.5), run_time=0.3)
            self.play(tool.animate.rotate(-0.5), run_time=0.3)

        text2 = Text("By day, just an ordinary man...", font_size=28).to_edge(UP)
        self.play(Write(text2))
        self.wait(2)
        self.play(FadeOut(text2))

        # -----------------------------
        # Scene 3: Transformation Hint
        # -----------------------------
        glow = Circle(radius=0.6, color=BLUE, fill_opacity=0.3).move_to(mechanic)

        self.play(FadeIn(glow))
        self.play(glow.animate.scale(1.5).set_opacity(0.1), run_time=2)

        text3 = Text("But when needed...", font_size=28).to_edge(UP)
        self.play(Write(text3))
        self.wait(2)
        self.play(FadeOut(text3), FadeOut(glow))

        # -----------------------------
        # Scene 4: Train Disaster
        # -----------------------------
        bridge = Line(LEFT * 4, RIGHT * 4).shift(UP * 1)
        train = Rectangle(width=3, height=0.5, color=RED).move_to(bridge.get_center())

        self.play(Create(bridge), FadeIn(train))

        # tilt train
        self.play(train.animate.rotate(0.3).shift(RIGHT * 2), run_time=1)

        # shaking effect
        self.play(Wiggle(train), run_time=1)

        text4 = Text("Disaster strikes.", font_size=28).to_edge(UP)
        self.play(Write(text4))
        self.wait(2)
        self.play(FadeOut(text4))

        # -----------------------------
        # Scene 5: Hero Entry
        # -----------------------------
        streak = Line(LEFT * 5, ORIGIN, color=WHITE)

        self.play(Create(streak), run_time=0.5)
        self.play(FadeOut(streak))

        hero = Circle(radius=0.4, color=WHITE).move_to(DOWN * 2)

        self.play(FadeIn(hero, scale=1.5))

        text5 = Text("Skyward.", font_size=32).to_edge(UP)
        self.play(Write(text5))
        self.wait(1.5)
        self.play(FadeOut(text5))

        # -----------------------------
        # Scene 6: The Rescue
        # -----------------------------
        self.play(hero.animate.move_to(train.get_bottom() + DOWN * 0.5))

        # lift train slowly
        self.play(train.animate.rotate(-0.3).shift(LEFT * 2 + UP * 0.5),
                  run_time=3, rate_func=smooth)

        text6 = Text("Not for glory... but to save lives.", font_size=28).to_edge(UP)
        self.play(Write(text6))
        self.wait(2)
        self.play(FadeOut(text6))

        # -----------------------------
        # Scene 7: Child Moment
        # -----------------------------
        child = Circle(radius=0.2, color=PINK).move_to(train.get_center())

        self.play(FadeIn(child))
        self.play(child.animate.shift(UP * 0.3), run_time=1)

        text7 = Text("He wouldn’t let anyone fall.", font_size=28).to_edge(UP)
        self.play(Write(text7))
        self.wait(2)
        self.play(FadeOut(text7), FadeOut(child))

        # -----------------------------
        # Scene 8: Hero Leaves
        # -----------------------------
        self.play(hero.animate.shift(UP * 4), run_time=2)
        self.play(FadeOut(hero))

        # -----------------------------
        # Scene 9: Ending
        # -----------------------------
        final_text = Text(
            "Not all heroes are seen...\nBut they are always there.",
            font_size=30
        )

        self.play(Write(final_text))
        self.wait(3)
````

````python
from manim import *

class TreeVisualization(Scene):
    def construct(self):
        # Title
        title = Text("Understanding Tree Nodes", font_size=36)
        self.play(Write(title))
        self.play(title.animate.to_edge(UP))

        # Create nodes (circles with labels)
        def create_node(label):
            circle = Circle(radius=0.4)
            text = Text(label, font_size=24)
            return VGroup(circle, text)

        # Nodes
        n1 = create_node("1")
        n2 = create_node("2")
        n3 = create_node("3")
        n4 = create_node("4")
        n5 = create_node("5")

        # Position nodes (tree structure)
        n1.move_to(UP * 2)
        n2.move_to(LEFT * 2 + UP * 0.5)
        n3.move_to(RIGHT * 2 + UP * 0.5)
        n4.move_to(LEFT * 3 + DOWN * 1.5)
        n5.move_to(LEFT * 1 + DOWN * 1.5)

        # Edges (lines)
        l1 = Line(n1.get_bottom(), n2.get_top())
        l2 = Line(n1.get_bottom(), n3.get_top())
        l3 = Line(n2.get_bottom(), n4.get_top())
        l4 = Line(n2.get_bottom(), n5.get_top())

        # Draw tree
        self.play(FadeIn(n1))
        self.play(FadeIn(n2), FadeIn(n3))
        self.play(Create(l1), Create(l2))
        self.play(FadeIn(n4), FadeIn(n5))
        self.play(Create(l3), Create(l4))

        self.wait(1)

        # Highlight Root
        root_text = Text("Root Node", color=YELLOW).next_to(n1, RIGHT)
        self.play(n1[0].animate.set_fill(YELLOW, opacity=0.5))
        self.play(Write(root_text))
        self.wait(1)

        # Highlight Inner Node
        inner_text = Text("Inner Node", color=BLUE).next_to(n2, LEFT)
        self.play(n2[0].animate.set_fill(BLUE, opacity=0.5))
        self.play(Write(inner_text))
        self.wait(1)

        # Highlight Leaf Nodes
        leaf_text = Text("Leaf Nodes", color=GREEN).to_edge(DOWN)
        self.play(
            n3[0].animate.set_fill(GREEN, opacity=0.5),
            n4[0].animate.set_fill(GREEN, opacity=0.5),
            n5[0].animate.set_fill(GREEN, opacity=0.5),
        )
        self.play(Write(leaf_text))

        self.wait(2)

        # Final message
        final_text = Text("Classify nodes based on Parent & Children", font_size=28)
        self.play(Transform(title, final_text))
        self.wait(2)
````

````python
from manim import *

class TreeSQLVisualization(Scene):
    def construct(self):

        # ------------------ TITLE ------------------
        title = Text("Tree Node Classification using SQL", font_size=34)
        self.play(Write(title))
        self.play(title.animate.to_edge(UP))

        # ------------------ TABLE ------------------
        table = Table(
            [["1", "NULL"],
             ["2", "1"],
             ["3", "1"],
             ["4", "2"],
             ["5", "2"]],
            col_labels=[Text("id"), Text("p_id")],
            include_outer_lines=True,
        ).scale(0.5)

        table.to_edge(LEFT)
        self.play(Create(table))
        self.wait(1)

        # ------------------ TREE ------------------
        def node(label):
            return VGroup(Circle(radius=0.3), Text(label, font_size=20))

        n1 = node("1").move_to(RIGHT*3 + UP*2)
        n2 = node("2").move_to(RIGHT*2 + UP*0.5)
        n3 = node("3").move_to(RIGHT*4 + UP*0.5)
        n4 = node("4").move_to(RIGHT*1.5 + DOWN*1)
        n5 = node("5").move_to(RIGHT*2.5 + DOWN*1)

        l1 = Line(n1.get_bottom(), n2.get_top())
        l2 = Line(n1.get_bottom(), n3.get_top())
        l3 = Line(n2.get_bottom(), n4.get_top())
        l4 = Line(n2.get_bottom(), n5.get_top())

        self.play(FadeIn(n1))
        self.play(FadeIn(n2), FadeIn(n3), Create(l1), Create(l2))
        self.play(FadeIn(n4), FadeIn(n5), Create(l3), Create(l4))
        self.wait(1)

        # ------------------ ROOT ------------------
        root_text = Text("p_id IS NULL → Root", font_size=24).to_edge(DOWN)
        self.play(n1[0].animate.set_fill(YELLOW, opacity=0.5))
        self.play(Write(root_text))
        self.wait(1)

        # ------------------ CHILDREN LOGIC ------------------
        child_text = Text("Nodes appearing in p_id → have children", font_size=24).to_edge(DOWN)
        self.play(Transform(root_text, child_text))

        highlight_ids = [table.get_rows()[i][1] for i in range(1, 5)]
        self.play(*[cell.animate.set_fill(BLUE, opacity=0.5) for cell in highlight_ids])
        self.wait(1)

        # ------------------ LEAF ------------------
        leaf_text = Text("id NOT IN p_id → Leaf", font_size=24).to_edge(DOWN)
        self.play(Transform(root_text, leaf_text))

        self.play(
            n3[0].animate.set_fill(GREEN, opacity=0.5),
            n4[0].animate.set_fill(GREEN, opacity=0.5),
            n5[0].animate.set_fill(GREEN, opacity=0.5),
        )
        self.wait(1)

        # ------------------ INNER ------------------
        inner_text = Text("Else → Inner", font_size=24).to_edge(DOWN)
        self.play(Transform(root_text, inner_text))

        self.play(n2[0].animate.set_fill(BLUE, opacity=0.5))
        self.wait(1)

        # ------------------ SQL QUERY ------------------
        sql_code = Code(
                        code_string="""
                                    SELECT id,
                                    CASE
                                        WHEN p_id IS NULL THEN 'Root'
                                        WHEN id NOT IN (SELECT p_id FROM Tree WHERE p_id IS NOT NULL) THEN 'Leaf'
                                        ELSE 'Inner'
                                    END AS type
                                    FROM Tree;
                                    """,
                                        language="sql",
                                        background="window",
                                    ).scale(0.5)

        sql_code.to_edge(DOWN)
        self.play(Write(sql_code))
        self.wait(2)

        # ------------------ FINAL ------------------
        final_text = Text("Classify using Parent + Children logic", font_size=28)
        self.play(Transform(title, final_text))
        self.wait(2)
````

````python
from manim import *

class TreeSQLStepByStep(Scene):
    def construct(self):

        # ---------------- TITLE ----------------
        title = Text("SQL Tree Node Classification", font_size=34)
        self.play(Write(title))
        self.play(title.animate.to_edge(UP))

        # ---------------- TABLE ----------------
        table = Table(
            [["1", "NULL"],
             ["2", "1"],
             ["3", "1"],
             ["4", "2"],
             ["5", "2"]],
            col_labels=[Text("id"), Text("p_id")],
            include_outer_lines=True,
        ).scale(0.5)

        table.to_edge(LEFT)
        self.play(Create(table))

        # ---------------- TREE ----------------
        def node(label):
            return VGroup(Circle(radius=0.3), Text(label, font_size=20))

        n1 = node("1").move_to(RIGHT*3 + UP*2)
        n2 = node("2").move_to(RIGHT*2 + UP*0.5)
        n3 = node("3").move_to(RIGHT*4 + UP*0.5)
        n4 = node("4").move_to(RIGHT*1.5 + DOWN*1)
        n5 = node("5").move_to(RIGHT*2.5 + DOWN*1)

        l1 = Line(n1.get_bottom(), n2.get_top())
        l2 = Line(n1.get_bottom(), n3.get_top())
        l3 = Line(n2.get_bottom(), n4.get_top())
        l4 = Line(n2.get_bottom(), n5.get_top())

        self.play(FadeIn(n1))
        self.play(FadeIn(n2), FadeIn(n3), Create(l1), Create(l2))
        self.play(FadeIn(n4), FadeIn(n5), Create(l3), Create(l4))

        self.wait(1)

        # ---------------- STEP TEXT ----------------
        step_text = Text("", font_size=24).to_edge(DOWN)
        self.play(FadeIn(step_text))

        # ---------------- STEP 1: ROOT ----------------
        self.play(step_text.animate.become(
            Text("Step 1: p_id IS NULL → Root", font_size=24).to_edge(DOWN)
        ))

        root_row = table.get_rows()[0]  # row with id=1
        self.play(root_row[1].animate.set_fill(YELLOW, opacity=0.7))
        self.play(n1[0].animate.set_fill(YELLOW, opacity=0.7))
        self.wait(2)

        # ---------------- STEP 2: p_id COLUMN ----------------
        self.play(step_text.animate.become(
            Text("Step 2: p_id column → Nodes with children", font_size=24).to_edge(DOWN)
        ))

        p_id_cells = [table.get_rows()[i][1] for i in range(1, 5)]
        self.play(*[cell.animate.set_fill(BLUE, opacity=0.5) for cell in p_id_cells])
        self.wait(2)

        # ---------------- STEP 3: LEAF ----------------
        self.play(step_text.animate.become(
            Text("Step 3: id NOT IN p_id → Leaf", font_size=24).to_edge(DOWN)
        ))

        self.play(
            n3[0].animate.set_fill(GREEN, opacity=0.7),
            n4[0].animate.set_fill(GREEN, opacity=0.7),
            n5[0].animate.set_fill(GREEN, opacity=0.7),
        )
        self.wait(2)

        # ---------------- STEP 4: INNER ----------------
        self.play(step_text.animate.become(
            Text("Step 4: Remaining → Inner", font_size=24).to_edge(DOWN)
        ))

        self.play(n2[0].animate.set_fill(ORANGE, opacity=0.7))
        self.wait(2)

        # ---------------- FINAL MESSAGE ----------------
        final_text = Text("SQL = Parent Check + Children Check", font_size=28)
        self.play(Transform(title, final_text))
        self.wait(2)
````

````python
from manim import *

class TripsSQLVisualization(Scene):
    def construct(self):

        # ---------------- TITLE ----------------
        title = Text("Understanding Cancellation Rate in SQL", font_size=34)
        self.play(Write(title))
        self.play(title.animate.to_edge(UP))

        # ---------------- MAIN TEXT HOLDER ----------------
        main_text = Text("", font_size=28).to_edge(DOWN)
        self.play(FadeIn(main_text))

        # ---------------- SCENE 1 ----------------
        self.play(Transform(main_text, Text(
            "We have two tables: Trips and Users", font_size=28
        ).to_edge(DOWN)))
        self.wait(2)

        # ---------------- TABLES ----------------
        trips = Table(
            [["1","1","10","completed","2013-10-01"],
             ["2","2","11","cancelled","2013-10-01"]],
            col_labels=[
                Text("id"), Text("client_id"),
                Text("driver_id"), Text("status"), Text("date")
            ],
        ).scale(0.5).to_edge(LEFT)

        users = Table(
            [["1","No"],["2","Yes"],["10","No"]],
            col_labels=[Text("id"), Text("banned")]
        ).scale(0.5).to_edge(RIGHT)

        self.play(Create(trips), Create(users))
        self.wait(2)

        # ---------------- STEP: JOIN ----------------
        self.play(Transform(main_text, Text(
            "Step 1: Join Users table twice (client + driver)", font_size=28
        ).to_edge(DOWN)))
        self.wait(2)

        # ---------------- STEP: FILTER ----------------
        self.play(Transform(main_text, Text(
            "Step 2: Remove banned users", font_size=28
        ).to_edge(DOWN)))

        # fade banned row
        self.play(users.get_rows()[1].animate.set_opacity(0.2))
        self.wait(2)

        # ---------------- STEP: CASE ----------------
        self.play(Transform(main_text, Text(
            "Step 3: Create is_cancelled column", font_size=28
        ).to_edge(DOWN)))
        self.wait(2)

        # ---------------- STEP: GROUP ----------------
        self.play(Transform(main_text, Text(
            "Step 4: Group by date", font_size=28
        ).to_edge(DOWN)))
        self.wait(2)

        # ---------------- STEP: AGG ----------------
        self.play(Transform(main_text, Text(
            "Step 5: SUM(cancelled) / COUNT(*)", font_size=28
        ).to_edge(DOWN)))
        self.wait(2)

        # ---------------- STEP: RESULT ----------------
        self.play(Transform(main_text, Text(
            "Final Result: Cancellation Rate per day", font_size=28
        ).to_edge(DOWN)))

        result = Text(
"""2013-10-01 → 0.33
2013-10-02 → 0.00
2013-10-03 → 0.50""",
            font_size=30
        )

        self.play(Write(result))
        self.wait(3)

        # ---------------- FINAL ----------------
        self.play(Transform(main_text, Text(
            "SQL = Step-by-step transformation", font_size=28
        ).to_edge(DOWN)))
        self.wait(2)
````

````python
from manim import *

class UnderstandingSlope(Scene):
    def construct(self):
        self.title_scene()
        self.what_is_slope()
        self.intuitive_meaning()
        self.rise_and_run()
        self.slope_formula()
        self.positive_slope()
        self.negative_slope()
        self.zero_slope()
        self.undefined_slope()
        self.full_range()
        self.slope_intercept()
        self.compare_slopes()
        self.summary()

    # ---------------------------------------------------
    # Scene 0: Title
    # ---------------------------------------------------
    def title_scene(self):
        title = Text(
            "Understanding Slope\nThe Steepness and Direction of a Line",
            font_size=42
        )

        self.play(Write(title))
        self.wait(2)
        self.play(FadeOut(title))

    # ---------------------------------------------------
    # Scene 1: What is Slope?
    # ---------------------------------------------------
    def what_is_slope(self):
        axes = Axes(
            x_range=[-5,5],
            y_range=[-4,4],
            axis_config={"include_numbers": False}
        )

        line1 = axes.plot(lambda x: 0.5*x, color=BLUE)
        line2 = axes.plot(lambda x: x, color=GREEN)
        line3 = axes.plot(lambda x: 2*x, color=RED)

        title = Text("What Is Slope?", font_size=36).to_edge(UP)

        definition = Text(
            "Slope = Steepness of a Line",
            font_size=30
        ).to_edge(DOWN)

        self.play(Create(axes))
        self.play(Create(line1), Create(line2), Create(line3))
        self.play(Write(title))
        self.play(Write(definition))

        self.wait(3)

        self.play(
            FadeOut(axes),
            FadeOut(line1),
            FadeOut(line2),
            FadeOut(line3),
            FadeOut(title),
            FadeOut(definition)
        )

    # ---------------------------------------------------
    # Scene 2: Intuitive Meaning
    # ---------------------------------------------------
    def intuitive_meaning(self):

        axes = Axes(
            x_range=[-4,4],
            y_range=[-3,5]
        )

        gentle = axes.plot(lambda x: 0.5*x, color=BLUE)
        medium = axes.plot(lambda x: x, color=GREEN)
        steep = axes.plot(lambda x: 2*x, color=RED)
        flat = axes.plot(lambda x: 0*x, color=YELLOW)

        labels = VGroup(
            Text("Small Slope", font_size=24),
            Text("Medium Slope", font_size=24),
            Text("Large Slope", font_size=24),
            Text("Zero Slope", font_size=24),
        ).arrange(DOWN).to_edge(RIGHT)

        self.play(Create(axes))
        self.play(Create(gentle))
        self.play(Create(medium))
        self.play(Create(steep))
        self.play(Create(flat))
        self.play(Write(labels))

        self.wait(3)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 3: Rise and Run
    # ---------------------------------------------------
    def rise_and_run(self):

        plane = NumberPlane()

        p1 = Dot([-2,-1,0], color=BLUE)
        p2 = Dot([2,2,0], color=BLUE)

        line = Line(p1.get_center(), p2.get_center())

        rise = Arrow(
            start=[2,-1,0],
            end=[2,2,0],
            color=GREEN
        )

        run = Arrow(
            start=[-2,-1,0],
            end=[2,-1,0],
            color=RED
        )

        rise_text = Text("Rise", font_size=30).next_to(rise, RIGHT)
        run_text = Text("Run", font_size=30).next_to(run, DOWN)

        formula = MathTex(
            r"\text{Slope}=\frac{\text{Rise}}{\text{Run}}"
        ).to_edge(UP)

        self.play(Create(plane))
        self.play(Create(line))
        self.play(FadeIn(p1,p2))

        self.play(Create(run))
        self.play(Write(run_text))

        self.play(Create(rise))
        self.play(Write(rise_text))

        self.play(Write(formula))

        self.wait(3)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 4: Formula
    # ---------------------------------------------------
    def slope_formula(self):

        formula = MathTex(
            r"m=\frac{y_2-y_1}{x_2-x_1}"
        ).scale(1.5)

        example = MathTex(
            r"m=\frac{4-2}{5-1}=\frac{2}{4}=\frac12"
        ).next_to(formula, DOWN)

        points = Text(
            "A(1,2), B(5,4)",
            font_size=32
        ).to_edge(UP)

        self.play(Write(points))
        self.play(Write(formula))
        self.play(Write(example))

        self.wait(4)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 5: Positive Slope
    # ---------------------------------------------------
    def positive_slope(self):

        axes = Axes()

        line = axes.plot(lambda x: x)

        title = Text("Positive Slope", font_size=36).to_edge(UP)

        formula = MathTex("m > 0").to_edge(DOWN)

        examples = Text(
            "m = 1   m = 2   m = 5",
            font_size=30
        ).next_to(formula, DOWN)

        self.play(Create(axes))
        self.play(Create(line))
        self.play(Write(title))
        self.play(Write(formula))
        self.play(Write(examples))

        self.wait(3)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 6: Negative Slope
    # ---------------------------------------------------
    def negative_slope(self):

        axes = Axes()

        line = axes.plot(lambda x: -x)

        title = Text("Negative Slope").to_edge(UP)

        formula = MathTex("m < 0")

        examples = Text(
            "m = -1   m = -3   m = -10"
        ).next_to(formula, DOWN)

        self.play(Create(axes))
        self.play(Create(line))
        self.play(Write(title))
        self.play(Write(formula))
        self.play(Write(examples))

        self.wait(3)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 7: Zero Slope
    # ---------------------------------------------------
    def zero_slope(self):

        axes = Axes()

        line = Line(
            LEFT*4,
            RIGHT*4,
            color=YELLOW
        )

        formula = MathTex("m=0")

        example = MathTex("y=3").next_to(formula, DOWN)

        self.play(Create(axes))
        self.play(Create(line))
        self.play(Write(formula))
        self.play(Write(example))

        self.wait(3)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 8: Undefined Slope
    # ---------------------------------------------------
    def undefined_slope(self):

        axes = Axes()

        line = Line(
            DOWN*3,
            UP*3,
            color=RED
        )

        formula = MathTex(
            r"m=\frac{\text{rise}}{0}"
        )

        example = MathTex("x=4").next_to(formula, DOWN)

        self.play(Create(axes))
        self.play(Create(line))
        self.play(Write(formula))
        self.play(Write(example))

        self.wait(3)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 9: Full Range
    # ---------------------------------------------------
    def full_range(self):

        slope_tracker = ValueTracker(-3)

        axes = Axes()

        def line_func():
            m = slope_tracker.get_value()
            return axes.plot(
                lambda x: m*x,
                x_range=[-3,3],
                color=BLUE
            )

        line = always_redraw(line_func)

        self.add(axes, line)

        self.play(
            slope_tracker.animate.set_value(-0.5),
            run_time=2
        )

        self.play(
            slope_tracker.animate.set_value(0),
            run_time=2
        )

        self.play(
            slope_tracker.animate.set_value(0.5),
            run_time=2
        )

        self.play(
            slope_tracker.animate.set_value(3),
            run_time=2
        )

        numberline = NumberLine(
            x_range=[-5,5,1]
        ).to_edge(DOWN)

        self.play(Create(numberline))

        self.wait(2)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 10: y = mx + b
    # ---------------------------------------------------
    def slope_intercept(self):

        equation = MathTex(
            r"y = mx + b"
        ).scale(2)

        slope_label = Text(
            "m → Controls Steepness",
            font_size=28
        ).next_to(equation, DOWN)

        intercept_label = Text(
            "b → Controls Position",
            font_size=28
        ).next_to(slope_label, DOWN)

        self.play(Write(equation))
        self.play(Write(slope_label))
        self.play(Write(intercept_label))

        self.wait(4)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 11: Compare Slopes
    # ---------------------------------------------------
    def compare_slopes(self):

        axes = Axes()

        l1 = axes.plot(lambda x: 0.5*x, color=BLUE)
        l2 = axes.plot(lambda x: x, color=GREEN)
        l3 = axes.plot(lambda x: 2*x, color=YELLOW)
        l4 = axes.plot(lambda x: 4*x, color=RED)

        legend = VGroup(
            MathTex("y=0.5x"),
            MathTex("y=x"),
            MathTex("y=2x"),
            MathTex("y=4x")
        ).arrange(DOWN).to_edge(RIGHT)

        self.play(Create(axes))
        self.play(
            Create(l1),
            Create(l2),
            Create(l3),
            Create(l4)
        )
        self.play(Write(legend))

        self.wait(4)

        self.play(*[FadeOut(m) for m in self.mobjects])

    # ---------------------------------------------------
    # Scene 12: Summary
    # ---------------------------------------------------
    def summary(self):

        title = Text("Summary", font_size=42).to_edge(UP)

        formula = MathTex(
            r"m=\frac{y_2-y_1}{x_2-x_1}"
        )

        table = MathTable(
            [
                ["Positive", "m = 2"],
                ["Negative", "m = -2"],
                ["Zero", "m = 0"],
                ["Undefined", "Vertical Line"]
            ],
            col_labels=[
                Text("Type"),
                Text("Example")
            ]
        ).scale(0.7)

        table.next_to(formula, DOWN)

        self.play(Write(title))
        self.play(Write(formula))
        self.play(Create(table))

        self.wait(5)
````
